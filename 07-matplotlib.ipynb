```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
```

## <font color=skyblue>_**plt.plot( <font color=orange>data</font>, <font color=orange>color</font>, <font color=orange>linestyle</font>, <font color=orange>marker</font>, <font color=orange>label</font>, <font color=orange>drawstyle</font> )**_</font>

```python
data = np.arange(10)
plt.plot(data)
```

### _**<font color=orange>color</font>, <font color=orange>linestyle</font>, <font color=orange>marker</font>, <font color=orange>label</font>**_ : 颜色, 标记, 线型, 图例

<font size=2><font color=skyblue>_**plt.subplots( <font color=orange>nrows</font>, <font color=orange>ncols</font>, <font color=orange>sharex</font>, <font color=orange>sharey</font> )**_</font>: 调整图片大小</font> 

```python
data = np.random.randn(30).cumsum()
plt.plot(data, color='b', linestyle='--', marker='o')
plt.plot(-data, 'ro--')
plt.subplots_adjust(top=0.6)
```

### _**<font color=orange>drawstyle</font>**_ : 

<font size=2><font color=skyblue>_**plt.legend( <font color=orange>loc</font> )**_</font>: 设置图例</font> 

```python
data = np.random.randn(30).cumsum()
plt.plot(data, 'k--', label='Default')
plt.plot(data, 'k-', drawstyle='steps-post', label='steps-post')
plt.legend(loc='best')
```

# Figure 和 Subplot

## <font color=skyblue>_**plt.figure( <font color=orange>figsize</font> )**_</font> : 创建一个的 Figure 对象; matplotlib 的图像都位于 Figure 对象中

```python
fig = plt.figure() #创建一个的 Figure 对象
```

## <font color=skyblue>_**<font color=orange>fig</font>.add_subplot( <font color=orange>nrows</font>, <font color=orange>ncols</font>, <font color=orange>index</font> )**_</font> : 创建一个的 AxesSubplot 对象, 位于 _[ nrows, ncols ]_ 的 _index_ 位置

```python
fig = plt.figure() #创建一个的 Figure 对象
ax1 = fig.add_subplot(2, 2, 1)
ax2 = fig.add_subplot(2, 2, 2)
ax3 = fig.add_subplot(2, 2, 3)
ax4 = fig.add_subplot(2, 2, 4)
data = np.random.randn(100)
plt.plot(data.cumsum(), 'k--')
ax1.plot(data.cumsum())
ax2.hist(data, bins=12, color='k', alpha=0.3) # 统计直方图
ax3.scatter(np.arange(100), data.cumsum()) # 散点图
plt.savefig('normal.png', dpi=1080, bbox_inches='tight')
```

## <font color=skyblue>_**plt.subplots( <font color=orange>nrows</font>, <font color=orange>ncols</font>, <font color=orange>sharex</font>, <font color=orange>sharey</font> )**_</font> : 创建 Figure 对象和 一组 AxesSubplot 对象（存放于 ndarray 中）

```python
fig, axes = plt.subplots(2, 1, sharex='all', sharey='all') # 共享x轴, y轴
# fig : Figure 对象 ; axes : ndarray 对象, 用于索引 AxesSubplot
plt.plot(data, 'k--')
axes[0].plot(data) #对图片进行索引
```

## <font color=skyblue>_**plt.subplots_adjust( <font color=orange>left</font>, <font color=orange>bottom</font>, <font color=orange>right</font>, <font color=orange>top</font>, <font color=orange>wspace</font>, <font color=orange>hspace</font> )**_</font> : 调整 subplot 大小和间距

```python
fig, axes = plt.subplots(2, 2, sharex=True, sharey=True)
for i in range(2):
    for j in range(2):
        axes[i, j].hist(np.random.randn(500), bins=50, color='k' ,alpha=0.5)
plt.subplots_adjust(left=0, bottom=0, right=2, top=1.2, 
                    wspace=0, hspace=0)
```

# 刻度, 标签 和 图例

```python
data = np.random.randn(500)
_ = plt.hist(data, bins=50, color='k' ,alpha=0.5)
```

## <font color=skyblue>_**plt.xlim( ), plt.xlim( <font color=orange>[ start, stop]</font> )**_</font> : 控制绘图范围

```python
plt.hist(data, bins=50, color='k' ,alpha=0.5)
plt.xlim(), plt.xlim([-2, 2])
```

## <font color=skyblue>_**<font color=orange>ax</font>.get_xlim( ), <font color=orange>ax</font>.set_xlim( <font color=orange>[ start, stop]</font> )**_</font> : 控制绘图范围

```python
fig, ax = plt.subplots(1, 1)
ax.hist(data, bins=50, color='k' ,alpha=0.5)
ax.get_xlim(), ax.set_xlim([-2,2])
```

## <font color=skyblue>_**plt.xticks( ), plt.xticks( <font color=orange>ticks</font>, <font color=orange>labels</font> )**_</font> : 控制刻度位置和标签

```python
data = np.random.randn(500)
plt.hist(data, bins=50, color='k' ,alpha=0.5)
plt.xticks(ticks = [-2, 0, 2] )
```

```python
data = np.random.randn(500)
plt.hist(data, bins=50, color='k' ,alpha=0.5)
plt.xticks(ticks = [-2, 0, 2], labels=['negative', 'zero', 'positive'] )
```

## <font color=skyblue>_**<font color=orange>ax</font>.get_xticks( ), <font color=orange>ax</font>.set_xticks( <font color=orange>ticks</font> )**_</font> : 控制刻度位置

## <font color=skyblue>_**<font color=orange>ax</font>.get_xticklabels( ), <font color=orange>ax</font>.set_xticklabels( <font color=orange>list</font> )**_</font> : 控制刻度标签

# 添加文本

## _**<font color=skyblue><font color=orange>ax</font>.text( <font color=orange>x</font>, <font color=orange>y</font>, <font color=orange>str</font>, <font color=orange>rotation</font>, <font color=orange>ha</font>, <font color=orange>va</font>, <font color=orange>fontdict <font color=ff8888>=</font> <font color=skyblue>dict( <font color=orange>...</font> )</font></font>, <font color=orange>bbox <font color=ff8888>=</font> <font color=skyblue>dict( <font color=orange>...</font> )</font></font> )</font>**_  
<font size=3> _**<font color=orange>fontdict <font color=ff8888>=</font> <font color=skyblue>dict( <font color=orange>fontsize</font>, <font color=orange>color</font>, <font color=orange>family</font>, <font color=orange>weight</font> )</font>**_ </font>  
<font size=3> _**<font color=orange>bbox <font color=ff8888>=</font> <font color=skyblue>dict( <font color=orange>facecolor</font>, <font color=orange>edgecolor</font>, <font color=orange>edgewidth</font>, <font color=orange>alpha</font>, <font color=orange>pad</font> )</font>**_ </font>  

> <font size=3> _**<font color=orange>rotation</font>**_ : 旋转 , _**<font color=orange>ha</font>**_ : 左右对齐方式 , _**<font color=orange>va</font>**_ : 上下对齐方式 </font>  
>  
> <font size=3> _**<font color=orange>fontdict</font>**_ : 字体属性字典</font>  
> <font size=3> _**<font color=orange>family</font>**_ : 字体 , _**<font color=orange>weight</font>**_ : 磅值</font>  
>  
> <font size=3> _**<font color=orange>bbox</font>**_ : 边框填充属性字典</font>  
> <font size=3> _**<font color=orange>facecolor</font>**_ : 填充色 , _**<font color=orange>edgecolor</font>**_ : 边框色, _**<font color=orange>edgewidth</font>**_ : 边框粗细</font></font>  
> <font size=3>  _**<font color=orange>alpha</font>**_ : 透明度 , _**<font color=orange>pad</font>**_ : 文本与框周围距离, _**<font color=orange>boxstyle</font>**_ : 形状参数</font></font>  

```python
fig, ax = plt.subplots(1, 1)
ax.hist(data, bins=50, color='k' ,alpha=0.5, label='one')
ax.text(0, 10, 'Hello world!', ha = 'center', va = 'bottom', 
        fontdict = dict(family='monospace', fontsize=20), 
        bbox = dict(facecolor='#74c476', edgecolor='b', alpha=0.7))
```

## _**<font color=skyblue><font color=orange>ax</font>.annotate( <font color=orange>text</font>, <font color=orange>xy</font>, <font color=orange>xytext</font>, <font color=orange>xycoords</font>, <font color=orange>textcoords</font>, <font color=orange>ha</font>, <font color=orange>va</font>, <font color=orange>weight</font>, <font color=orange>color</font>, <font color=orange>bbox</font>, <font color=orange>arrowprops</font> )</font>**_  
<font size=3> _**<font color=orange>xy <font color=ff8888>=</font> <font color=skyblue>( <font color=orange>x</font>, <font color=orange>y</font> )</font>**_ </font>  
<font size=3> _**<font color=orange>xytext <font color=ff8888>=</font> <font color=skyblue>( <font color=orange>x</font>, <font color=orange>y</font> )</font>**_ </font>  
<font size=3> _**<font color=orange>bbox <font color=ff8888>=</font> <font color=skyblue>dict( <font color=orange>facecolor</font>, <font color=orange>edgecolor</font>, <font color=orange>edgewidth</font>, <font color=orange>alpha</font>, <font color=orange>pad</font> )</font>**_ </font>  
<font size=3> _**<font color=orange>arrowprops <font color=ff8888>=</font> <font color=skyblue>dict( <font color=orange>facecolor</font>, <font color=orange>hatch</font>, <font color=orange>alpha</font>, <font color=orange>shrink</font>, <font color=orange>width</font>, <font color=orange>headwidth</font>, <font color=orange>headlength</font> )</font>**_ </font>  

> <font size=3> _**<font color=orange>xy</font>**_ : 箭头指向位置 , _**<font color=orange>xytext</font>**_ : 文本位置 </font>  
> <font size=3> _**<font color=orange>xycoords</font>**_ : 箭头位置的参考坐标系 , _**<font color=orange>textcoords</font>**_ : 选择文本位置的参考坐标系 </font>  
> <font size=3> _**<font color=orange>ha</font>**_ : 文本左右对齐方式 , _**<font color=orange>va</font>**_ : 文本上下对齐方式 </font>  
> <font size=3> _**<font color=orange>weight</font>**_ : 文本磅值 , _**<font color=orange>va</font>**_ : 文本颜色 </font>  
> <font size=3> _**<font color=orange>bbox</font>**_ : 文本边框填充属性字典 </font>  
>  
> <font size=3> _**<font color=orange>arrowprops</font>**_ : 箭头属性字典</font>  
> <font size=3> _**<font color=orange>facecolor</font>**_ : 填充色 , _**<font color=orange>hatch</font>**_ : 填充形状 </font>  
> <font size=3> _**<font color=orange>shrink</font>**_ : 两端收缩控制, _**<font color=orange>width</font>**_ : 箭身宽度</font>  
> <font size=3> _**<font color=orange>headwidth</font>**_ : 箭头宽度, _**<font color=orange>headlength</font>**_ : 箭头长度</font>  

```python
fig, ax = plt.subplots(1, 1)
ax.hist(data, bins=50, color='k' ,alpha=0.5, label='one')
ax.annotate('Hello World!', (0, 10), (2, 25), ha = 'center', va = 'top', fontsize=20, 
            arrowprops=dict(facecolor='k', width=0.1, headwidth=5, headlength=5, shrink=0.05), 
            bbox = dict(facecolor='#74c476', edgecolor='b', alpha=0.7)
            )
```

# 添加图形

## _**<font color=skyblue><font color=orange>ax</font>.add_patch( <font color=orange>shp</font> )</font>**_  

<font size=4> _**<font color=skyblue>plt.Rectangle( <font color=orange>xy</font>, <font color=orange>width</font>, <font color=orange>height</font>, <font color=orange>angle</font>, <font color=orange>color</font>, <font color=orange>alpha</font> )</font>**_  </font>  
<font size=4> _**<font color=skyblue>plt.Circle( <font color=orange>xy</font>, <font color=orange>r</font>, <font color=orange>color</font>, <font color=orange>alpha</font> )</font>**_  </font>  
<font size=4> _**<font color=skyblue>plt.Polygon( <font color=orange>[ xy1, xy2, ... ]</font>, <font color=orange>color</font>, <font color=orange>alpha</font> )</font>**_  </font>  

```python
fig = plt.figure(figsize=(4.5, 9.5))
ax = fig.add_subplot(1, 1, 1)
ax.set_xlim(0,9)
ax.set_ylim(1,20)
# 正方形
rect = plt.Rectangle((0, 1), 9, 20, color='#b7e464', alpha=1)
ax.add_patch(rect)
rect = plt.Rectangle((0.6, 13.9), 3.2, 5.5, color='#8db53c', alpha=1)
ax.add_patch(rect)
rect = plt.Rectangle((0.7, 14), 3, 5.3, color='#b7e464', alpha=1)
ax.add_patch(rect)
rect = plt.Rectangle((6.3, 0), 2, 20, color='k', alpha=1)
ax.add_patch(rect)
rect = plt.Rectangle((5.5, 0), 0.4, 20, color='k', alpha=1)
ax.add_patch(rect)
# 圆形
circ = plt.Circle((1.75, 15.15), 0.8, color='k', alpha=1)
ax.add_patch(circ)
circ = plt.Circle((1.75, 15.15), 0.25, color='#121233', alpha=1)
ax.add_patch(circ)
circ = plt.Circle((1.75, 18.15), 0.8, color='k', alpha=1)
ax.add_patch(circ)
circ = plt.Circle((1.75, 18.15), 0.25, color='#121233', alpha=1)
ax.add_patch(circ)
circ = plt.Circle((3.1, 16.65), 0.4, color='k', alpha=1)
ax.add_patch(circ)
circ = plt.Circle((3.1, 16.65), 0.25, color='#121233', alpha=1)
ax.add_patch(circ)
circ = plt.Circle((3.1, 15.15), 0.35, color='k', alpha=1)
ax.add_patch(circ)
circ = plt.Circle((3.1, 18.15), 0.35, color='k', alpha=1)
ax.add_patch(circ)
circ = plt.Circle((3.1, 15.15), 0.25, color='w', alpha=0.9)
ax.add_patch(circ)
circ = plt.Circle((3.1, 18.15), 0.25, color='w', alpha=0.9)
ax.add_patch(circ)
ax.text(7.3, 2.5, 'realme', ha = 'center', va = 'bottom', rotation=90, 
        fontdict = dict(fontsize=30, color='#74c476'))
for i in range(0, 9):
        ax.text(5.7, 1.4+i*2, 'DARE TO LEAP', ha = 'center', va = 'bottom', rotation=90, 
        fontdict = dict(fontsize=7, color='w'))
ax.text(5.7, 1.4+(i+1)*2, 'DAR', ha = 'center', va = 'bottom', rotation=90, 
        fontdict = dict(fontsize=7, color='w'))
ax.text(1.75, 16.75, '64MP', ha = 'center', va = 'center', alpha=0.9, 
        fontdict = dict(fontsize=16, color='#EEEEEE', weight='light'))
ax.text(1.75, 16.35, 'MATRIX CAMERA', ha = 'center', va = 'center', alpha=0.9, 
        fontdict = dict(fontsize=5, color='#EEEEEE', weight='light'))
```
