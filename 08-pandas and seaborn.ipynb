pandas and seaborn

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
%matplotlib inline
```

## _**<font color=skyblue><font color=orange>obj</font>.plot( <font color=orange>use_index</font>, <font color=orange>xlim</font>, <font color=orange>xticks</font>, <font color=orange>xlabel</font>, <font color=orange>rot</font>, <font color=orange>logy</font>, <font color=orange>figsize</font>, <font color=orange>label</font>, <font color=orange>legend</font>, <font color=orange>title</font>, <font color=orange>kind</font>, <font color=orange>style</font>, <font color=orange>alpha</font>, <font color=orange>grid</font>, <font color=ff8888>ax</font> )</font>**_ : 
> <font size=3> _**<font color=orange>use_index</font>**_ : 是否将对象索引用作刻度标签</font>  
> <font size=3> _**<font color=orange>xlim</font>**_ : x轴界限, _**<font color=orange>xticks</font>**_ : x轴刻度值, _**<font color=orange>xlabel</font>**_ : x轴名称</font>  
> <font size=3> _**<font color=orange>logy</font>**_ : 是否在y轴使用对数标尺（ _log(p) → h_ ）</font>  
> <font size=3> _**<font color=orange>label, legend</font>**_ : 图例</font>  
> <font size=3> _**<font color=orange>kind</font>**_ : 控制画图种类, 如: _'line', 'bar', 'barh', 'kde'_ </font>  
> <font size=3> _**<font color=orange>style</font>**_ : 控制画图属性, 如: _'ko--'_ </font>  
> <font size=3> _**<font color=orange>grid</font>**_ : 是否显示网格线</font>  
> <font size=3> _**<font color='ff8888'>ax**_ : 指定要在其上进行绘制的 _**subplot**_ 对象</font></font>  

```python
fig, axes = plt.subplots(1,1)
s = pd.Series(np.array([1,2,3,4]))
s.plot(kind='bar', grid=True, rot=0)
```

## _**<font color=skyblue><font color=orange>frame</font>.plot( <font color=orange>... ...</font>, <font color=orange>subplots</font>, <font color=orange>sharex</font>, <font color=orange>sharey</font>, <font color=orange>sortcolumns</font> )</font>**_ : 
> <font size=3> _**<font color=orange>subplots</font>**_ : 是否将 frame 各列单独绘制</font>  
> <font size=3> _**<font color=orange>sharex</font>**_ : _subplots=True_ : 是否共用一个 x 轴, _**<font color=orange>sharey</font>**_ :  _subplots=True_ : 是否共用一个 y 轴</font>  
> <font size=3> _**<font color=orange>sort_columns</font>**_ : 是否以字母表顺序绘制各列</font>  

```python
df = pd.DataFrame(np.random.randn(10, 4).cumsum(0),
                  columns=['A', 'B', 'C', 'D'],
                  index=np.arange(0, 100, 10))
df.plot()
```

```python
df.plot(subplots=True)
```

## _**<font color=skyblue><font color=orange>frame</font>.plot.line( <font color=orange>... ...</font> )</font>**_ : 折线图

```python
fig, axes = plt.subplots(2, 1)
fig.subplots_adjust()
data = pd.DataFrame(np.random.rand(6, 4), 
                    index=['one', 'two', 'three', 'four', 'five', 'six'],
                    columns=pd.Index(['A', 'B', 'C', 'D'], 
                    name='Genus'))
data.plot.bar(ax=axes[0])
data.plot.line(ax=axes[1])
```

## _**<font color=skyblue><font color=orange>frame</font>.plot.bar( <font color=orange>stacked</font>, <font color=orange>...</font> ), <font color=orange>frame</font>.plot.barh( <font color=orange>stacked</font>, <font color=orange>...</font> )</font>**_ : 条形图

```python
data = pd.DataFrame(np.arange(1,25).reshape(4, 6).T, 
                    index=['one', 'two', 'three', 'four', 'five', 'six'],
                    columns=pd.Index(['A', 'B', 'C', 'D'], 
                    name='Genus'))
data
```

```python
data.plot.bar()
data.plot.barh(stacked=True) #叠加
```

## _**<font color=skyblue>sns.barplot( <font color=orange>data</font>, <font color=orange>x</font>, <font color=orange>y</font>, <font color=orange>hue</font>, <font color=orange>orient</font> )</font>**_ : 平均值及置信度

```python
tips = pd.read_csv('pydata-book-2nd-edition/examples/tips.csv')
tips.head()
```

```python
tips['tip/total_bill'] = tips['tip'] / tips['total_bill']
sns.barplot(data=tips, x='tip/total_bill', y='day', orient='h')
```

## _**<font color=skyblue><font color=orange>frame</font>.plot.hist( <font color=orange>bins</font>, <font color=orange>...</font> ), <font color=orange>frame</font>.plot.density( <font color=orange>...</font> )</font>**_ : 直方图和密度图

```python
tips = pd.read_csv('pydata-book-2nd-edition/examples/tips.csv')
tips['tip/total_bill'] = tips['tip'] / tips['total_bill']
tips.head()
```

```python
tips['tip/total_bill'].plot.hist(bins=50)
```

```python
tips['tip/total_bill'].plot.density()
```

## _**<font color=skyblue>sns.regplot( <font color=orange>data</font>, <font color=orange>x</font>, <font color=orange>y</font>, <font color=orange>hue</font> )</font>**_ : 散点图

```python
macro = pd.read_csv('pydata-book-2nd-edition/examples/macrodata.csv')
data = macro[['cpi', 'm1', 'tbilrate', 'unemp']]
trans_data = np.log(data).diff().dropna()
trans_data.head()
sns.regplot(x = 'm1', y = 'unemp', data=trans_data)
```

## _**<font color=skyblue>sns.pairplot( <font color=orange>data</font>, <font color=orange>diag_kind</font>, <font color=orange>plot_kws</font> )</font>**_ : 散步图矩阵

```python
sns.pairplot(trans_data, diag_kind='kde')
```

## _**<font color=skyblue>sns.catplot( <font color=orange>data</font>, <font color=orange>x</font>, <font color=orange>y</font>, <font color=orange>hue</font>, <font color=orange>row</font>, <font color=orange>col</font>, <font color=orange>kind</font> )</font>**_ : 分面网格

```python
tips = pd.read_csv('pydata-book-2nd-edition/examples/tips.csv')
tips['tip_pct'] = tips['tip'] / tips['total_bill']
sns.catplot(x='day', y='tip_pct', hue='time', col='smoker',
               kind='bar', data=tips)
```

```python
sns.catplot(x='day', y='tip_pct', row='time',
               col='smoker',
               kind='bar', data=tips)
```

```python
sns.catplot(x='tip_pct', y='day', kind='bar',
               data=tips[tips.tip_pct < 0.5])
```

```python
sns.catplot(x='tip_pct', y='day', kind='box',
               data=tips[tips.tip_pct < 0.5])
```
